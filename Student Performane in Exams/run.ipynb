```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('StudentsPerformance.csv')
data.head()
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
data['total']=data['math score'] + data['reading score'] + data['writing score']
```

```python
data['average']=data['total']/3
```

```python
sns.distplot(data['math score'])
```

```python
sns.pairplot(data)
```

```python
sns.barplot(data['race/ethnicity'],data['average'])
```

```python
sns.barplot(data['parental level of education'],data['average'])
```

```python
sns.barplot(data['test preparation course'],data['average'])
```

```python
data['math_pass_status']=np.where(data['math score'] < 40,'F','P')
```

```python
data['read_pass_status']=np.where(data['reading score'] < 40,'F','P')
```

```python
data['write_pass_status']=np.where(data['writing score'] < 40,'F','P')
```

```python
data.head()
```

```python
data['math_pass_status'].value_counts()
```

```python
data['read_pass_status'].value_counts()
```

```python
p = sns.countplot(x='gender',data=data,hue='math_pass_status',palette='bright')
_ = plt.setp(p.get_xticklabels(),rotation=90)
```

```python
p = sns.countplot(x='parental level of education',data=data,hue='math_pass_status',palette='bright')
_ = plt.setp(p.get_xticklabels(),rotation=90)
```
